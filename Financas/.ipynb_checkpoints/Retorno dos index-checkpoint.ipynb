```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pandas_datareader import data as wb
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI']
mydata= pd.DataFrame()
for i in tickers:
    mydata[i] = wb.DataReader(i, data_source='yahoo', start='1997-1-1', end='2017-3-23')['Close']
```
