```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('data/iris.csv')
```

```python
data.head()
```

```python
previsores = data.iloc[:, 0:4].values
classe = data.iloc[:, 4].values
```

```python
labelEncoder = LabelEncoder()
classe = labelEncoder.fit_transform(classe)
classe_dummy = np_utils.to_categorical(classe)
```

```python
classe_dummy
```

```python
pre_train, pre_test, classe_train, classe_test = train_test_split(previsores, classe_dummy, test_size=0.25)
```

```python
classificador = Sequential()
classificador.add(Dense(units=4, activation='relu', input_dim=4))
classificador.add(Dense(units=4, activation='relu'))
classificador.add(Dense(units=3, activation='softmax'))
classificador.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
classificador.fit(pre_train, classe_train, batch_size=10, epochs=1000)
```

```python
resultado = classificador.evaluate(pre_test, classe_test)
```

```python
previsoes = classificador.predict(pre_test)
```

```python
previsoes = (previsoes>0.5)
```

```python
previsoes
```

```python
classe_test2 = [np.argmax(t) for t in classe_test]
classe_test2
previsoes2 = [np.argmax(t) for t in previsoes]
matriz = confusion_matrix(previsoes2, classe_test2)
```
